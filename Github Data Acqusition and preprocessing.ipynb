```python
userNames = []
users = []
url = f"https://api.github.com/users"
users_data = requests.get(url).json()
for user_data in users_data:
    userNames.append(user_data['login'])
    users.append(user_data)

print(userNames)
```

```python
from bs4 import BeautifulSoup

def find_languages_used(repo_info):
    languages=set()
    for j,y in enumerate(repo_info["repos"]):
        link=y["clone_url"]
        page=requests.get(link)
        git_page = BeautifulSoup(page.content, features='html.parser')
        for lang in git_page.find_all('ul',{"class":"list-style-none"}):
            b=lang.find_all("li",{"class":"d-inline"})
            for ae in b:
                languages.add(ae.find('span', {"class":"color-text-primary text-bold mr-1"}).get_text().strip())
    return languages
```

We have created database of users with github data.

On this data if Companies want to hire users based on git profile data. These are sample usecases:
- when companies wants to get users based on number of followers. Assumming that, user having followers based on his coding skills. So that they can check the languages in thier repositories and based on that they hire users

```python
def get_users_languages_based_on_followers(data,followers):
    data_new=[]
    for i,x in enumerate(data):
        dicts={}
        if x["followers"] > followers:
            languages=find_languages_used(x)
            dicts["username"]=x["login"]
            dicts["languages"]=list(languages)
            data_new.append(dicts)
    return pd.DataFrame(data_new)

user_lang=get_users_languages_based_on_followers(users_data_with_repo,3158)
user_lang
```

- When companies want to get users based on minimum number of repositories. Assuming that, users having coding skills based on the repositories. So that they can check languages for those users and hire them.

```python
def get_users_languages_based_on_repos(data,repo_n):
    users_data_with_repo_new=[]
    for i,x in enumerate(users_data_with_repo):
        dicts={}
        if len(x["repos"]) == repo_n:
            languages=find_languages_used(x)
            dicts["username"]=x["login"]
            dicts["languages"]=list(languages)
            users_data_with_repo_new.append(dicts)
    return pd.DataFrame(users_data_with_repo_new)

user_repo=get_users_languages_based_on_repos(users_info,60)
user_repo
```

- When companies want to hire user based on username or user profile link of github in thier resume.

```python
def get_user_profile_using_username_or_githuubLink(data,name_link):
    if "https" in name_link:
        github_profile=name_link
        username=''
    else:
        username=name_link
        github_profile=''
    users_data_with_repo_new=[]
    for i,x in enumerate(users_data_with_repo):
        dicts={}
        if username==x["login"] or github_profile==x["html_url"]:
            languages=find_languages_used(x)
            dicts["username"]=x["login"]
            dicts["languages"]=list(languages)
            users_data_with_repo_new.append(dicts)
    return pd.DataFrame(users_data_with_repo_new)

user_name_profile=get_user_profile_using_username_or_githuubLink(users_info,"https://github.com/defunkt")
user_name_profile
```

- When companies want to hire users based on followers, repositories and checking languages for those users and hiring based on this criteria 

```python
def get_users_languages_based_on_repos_and_followers(users_data_with_repo,repo_n,follow_n):
    users_data_with_repo_new=[]
    for i,x in enumerate(users_data_with_repo):
        dicts={}
        if len(x["repos"]) == repo_n and x["followers"]>follow_n:
            languages=find_languages_used(x)
            dicts["username"]=x["login"]
            dicts["languages"]=list(languages)
            users_data_with_repo_new.append(dicts)
    return pd.DataFrame(users_data_with_repo_new)

user_name_repo_followers=get_users_languages_based_on_repos_and_followers(users_info,60,500)
user_name_repo_followers
```
